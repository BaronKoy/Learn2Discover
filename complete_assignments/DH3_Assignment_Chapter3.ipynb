```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
img = mpimg.imread('data/rat_cerebellum.jpeg'); imgplot = plt.imshow(img)
```

```python
r = img[:, :, 0]
g = img[:, :, 1]
b = img[:, :, 2]
fig, ax = plt.subplots(ncols = 3)
ax[0].hist(r.flatten(), bins = 50, color = 'red'); ax[0].set_xlabel('Red contribution', color = 'red')
ax[1].hist(g.flatten(), bins = 50, color = 'green'); ax[1].set_xlabel('Green contribution', color= 'green')
ax[2].hist(b.flatten(), bins = 50, color ='blue'); ax[2].set_xlabel('blue contribution', color = 'blue')

fig, ax = plt.subplots(ncols = 3)
imgr = ax[0].imshow(r, cmap = 'Reds'); fig.colorbar(imgr, ax=ax[0], shrink=0.5)
imgb = ax[2].imshow(b, cmap = 'Blues'); fig.colorbar(imgb, ax=ax[2], shrink=0.5)
imgg = ax[1].imshow(g, cmap = 'Greens'); fig.colorbar(imgg, ax=ax[1], shrink=0.5)
```

```python
#Im not sure I understand this part? I had to check the answer in the datahandling 3 doc
rmask = r > 150; rmasked = r*rmask
gmask = g > 150; gmasked = g*gmask
bmask = b > 150; bmasked = b*bmask
#Changing these thresholds alters which pixel intensitys are included in the final Masked image.

fig, ax = plt.subplots(nrows=3, ncols=2)
ax[0, 0].imshow(rmask, cmap='gray'); ax[0, 0].set_title('Red binary mask')
ax[0, 1].imshow(rmasked, cmap='Reds'); ax[0, 1].set_title('Masked image')
ax[1, 0].imshow(gmask, cmap='gray'); ax[1, 0].set_title('Green binary mask')
ax[1, 1].imshow(gmasked, cmap='Greens'); ax[1, 1].set_title('Masked image')
ax[2, 0].imshow(bmask, cmap='gray'); ax[2, 0].set_title('Blue binary mask')
ax[2, 1].imshow(bmasked, cmap='Blues'); ax[2, 1].set_title('Masked image')
```

```python
import numpy as np
total = img.shape[0]*img.shape[1]
rcount = np.sum(rmask); gcount = np.sum(gmask); bcount = np.sum(bmask)
rdone = (rcount/total*100)
gdone = (gcount/total*100)
bdone = (bcount/total*100)
print('red=',rdone,'%', sep='')
print ('green=',gdone,'%',sep='')
print ('blue=',bdone,'%',sep='')
```
